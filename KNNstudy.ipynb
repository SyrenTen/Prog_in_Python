```python
class KNNClassifier:
    def __init__(self, all_data: pd.DataFrame, to_predict: str):
        train_test = train_test_split(all_data, test_size=0.1)
        self.train_data: pd.DataFrame = train_test[0]
        self.test_data: pd.DataFrame = train_test[1]
        self.to_predict: str = to_predict
        self.stats: dict = {}

        self.prepare()

    def standardize(self, data: dict | pd.DataFrame) -> None:
        """
        Standardizes an input data using internal stats (mean and standard deviation)
        """
        # TODO: Pour un attribut passé en paramètre, standardiser les données
        for column, stats in self.stats.items():
            data[column] = (data[column] - stats['mean']) / stats['std']

    def prepare(self) -> None:
        """
        Computes the distance between two entries
        """
        # TODO: Calculer et stocker la moyenne et l'écart type de chaque attribut 
        for column in self.train_data.columns:
            if column == self.to_predict:
                continue
            self.stats[column] = {
                "mean": self.train_data[column].mean(),
                "std": self.train_data[column].std(),
            }

        self.standardize(self.train_data)

    def distance(self, entry1: dict, entry2: dict) -> float:
        """
        Computes the distance between two entries
        """
        # TODO: Calculer la distance entre 2 entrées
        sum = 0
        for column in self.stats:
            sum = (entry1[column] - entry2[column])**2

        return np.sqrt(sum)

    def predict(self, entry: dict, k: int = 5) -> str:
        """
        Predicts the output class for a given entry
        """
        self.standardize(entry)

        # TODO: Trouver la classe majoritaire parmi les k plus proches voisins 

    def compute_accuracy(self, k: int = 5) -> float:
        """
        Compute accuracy over all test dataset
        """
        correct = 0

        # TODO: Calculer la justesse (accuracy) 

        return correct / len(self.test_data) * 100

knn = KNNClassifier(pd.read_csv("wine.csv"), "region")
```
