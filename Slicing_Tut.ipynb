```python
# Exercice 3
# Téléchargez lenna.png, puis chargez l’image dans ipython avec plt.imread, puis...
# Afficher l’image
image = plt.imread('lenna.png')
plt.imshow(image)
```

```python
# Extraire la partie centrale de l’image (un carré de 100×100 pixels) et l’afficher
image = plt.imread('lenna.png')
centerx, centery = [int(image.shape[0] / 2) - 50, int(image.shape[1] / 2) - 50]
image = image[centerx:centerx + 100, centery:centery + 100]
plt.imshow(image)
```

```python
# Remplacer un carré 50×50 pixels en haut à gauche par du noir.
image = plt.imread('lenna.png')
image[0:50,0:50] = 0
plt.imshow(image)
```

```python
h, w, c = image.shape
h, w, c
```

```python
# Modifier la couleur de la bande centrale en rouge
image = plt.imread('lenna.png')
image[256] = (255, 0, 0)
plt.imshow(image)
```

```python
# Effectuer une symétrie d’axe vertical (au centre de l’image).
image_sym = np.flip(image, axis=1)
plt.imshow(image_sym)
```

```python
# Faire pivoter l’image de 180°
image_rot180 = np.rot90(image, 2)
plt.imshow(image_rot180)
```

```python
# Augmenter la luminosité d’une zone
image_bright = image.copy()
image_bright[30:70, 30:70] = np.clip(image_bright[30:70, 30:70] + 50, 0, 255)
plt.imshow(image_bright)
```

```python
# Mettre la diagonale principale à zéro
image_diag0 = image.copy()
for i in range(min(image_diag0.shape[:2])):
    image_diag0[i, i] = 0
plt.imshow(image_diag0)
```

```python
# Convertir l’image en niveaux de gris
image_gray = np.dot(image[..., :3], [0.2989, 0.587, 0.114]).astype(np.uint8)
plt.imshow(image_gray)
```

```python
# Afficher l’image en négatif
image_neg = 255 - image
plt.imshow(image_neg)
```

```python
# Afficher les images correspondant aux canaux rouge, vert et bleu en même temps
image_red = image.copy()
image_red[:, :, 1:] = 0  # Supprime le vert et le bleu

image_green = image.copy()
image_green[:, :, [0, 2]] = 0  # Supprime le rouge et le bleu

image_blue = image.copy()
image_blue[:, :, :2] = 0  # Supprime le rouge et le vert
plt.imshow(image_red)
```

```python
plt.imshow(image_green)
```

```python
plt.imshow(image_blue)
```

```python
# Interlacer l’image de Lenna normale avec l’image de Lenna pivotée à 180°, une ligne sur deux
image_interlace = image.copy()
image_interlace[::2] = image_rot180[::2]
plt.imshow(image_interlace)
```
