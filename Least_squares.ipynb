```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import r2_score
```

```python
# On veut minimiser ||X w - y||^2
# w, r = np.linalg.lstsq(X, y)[:2]
```

```python
data1 = pd.read_csv('data_1.csv')
data1
```

```python
#Les fichiers data_1.csv, data_2.csv, data_3.csv, data_4.csv contiennent des donnÃ©es (x,y). Pour chacun dâ€™eux:
# Tracez le nuage de points reprÃ©sentant y en fonction de x.

xpoints = data1.x
ypoints = data1.y

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
X = np.vstack([data1.x, np.ones(len(data1.x))]).T # vstack - Stack arrays in sequence vertically (row wise)
w, r = np.linalg.lstsq(X, data1.y)[0] # Utilisez la fonction lstsq pour estimer les paramÃ¨tres
```

```python
# Proposez un modÃ¨le linÃ©aire y=f(x, w)
y_pred = w * data1.x + r
y_pred
```

```python
plt.plot(xpoints, ypoints, 'o')
plt.plot(data1.x, y_pred)
```

```python
# Calculez le coefficient de dÃ©termination R^2
rscore = r2_score(data1.y, y_pred)
rscore
```

```python
# or calculated by hand:
model = []
for i in data1.x:
    model.append(w*i+r)
model
```

```python
Spred = np.sum((model - data1.y)**2)
Spred
```

```python
Smean = np.sum((data1.y - np.mean(data1.y))**2)
Smean
```

```python
r2score = 1 - (Spred/Smean)
r2score
```
